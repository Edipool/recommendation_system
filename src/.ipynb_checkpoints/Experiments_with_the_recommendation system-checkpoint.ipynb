```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import datetime
import pickle
import re
import string

from nltk.stem import WordNetLemmatizer 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.compose import ColumnTransformer
from category_encoders import TargetEncoder
from category_encoders.one_hot import OneHotEncoder
from sklearn.metrics import roc_auc_score
from catboost import CatBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, f1_score, roc_curve, auc
from sklearn.metrics import RocCurveDisplay
```

## Reading locally saved dfs

```python
df_user_data = pd.read_csv("../datasets/df_user_data.csv", sep=';')
df_post_text = pd.read_csv("../datasets/df_post_text.csv", sep=';')
df_feed_data = pd.read_csv("../datasets/df_feed_data.csv", sep=';')
```

```python
### Let 's clean the data from action != view

df_feed_data = df_feed_data[df_feed_data.action == 'view']

df_feed_data.head()
```

```python
wnl = WordNetLemmatizer()

def preprocessing(line, token=wnl):
    line = line.lower()
    line = re.sub(r"[{}]".format(string.punctuation), " ", line)
    line = line.replace('\n\n', ' ').replace('\n', ' ')
    line = ' '.join([token.lemmatize(x) for x in line.split(' ')])
    return line


tfidf = TfidfVectorizer(
    stop_words='english',
    preprocessor=preprocessing
)
```

```python
tfidf_data = (
    tfidf
    .fit_transform(df_post_text['text'])
    .toarray()
)

tfidf_data
```

```python
tfidf_data = pd.DataFrame(
    tfidf_data,
    index=df_post_text.post_id,
    columns=tfidf.get_feature_names_out()
)

tfidf_data
```

```python
### Generate new features based on Tf Idf

df_post_text['TotalTfIdf'] = tfidf_data.sum(axis=1).reset_index()[0]
df_post_text['MaxTfIdf'] = tfidf_data.max(axis=1).reset_index()[0]
df_post_text['MeanTfIdf'] = tfidf_data.mean(axis=1).reset_index()[0]

df_post_text.head()
```

### Let's combine the date frames

```python
### Let's reproduce the dataframe with all the new features

df = pd.merge(df_feed_data,
              df_post_text,
              on='post_id',
              how='left')

df.head()
```

```python
df = pd.merge(df,
              df_user_data,
              on='user_id',
              how='left')

df.head()
```

```python
### The time of the recommendation is very important. 
### Therefore, we will highlight the features.

df['hour'] = pd.to_datetime(df['timestamp']).apply(lambda x: x.hour)
df['month'] = pd.to_datetime(df['timestamp']).apply(lambda x: x.month)

df.head()
```

```python
### Remove all unnecessary columns

df = df.drop([
    'action',
    'text',
],
    axis=1)
```

## Let's look at the correlations of signs

```python
### Let's build a correlation matrix
data_corr = df.corr()

colormap = plt.cm.viridis
plt.figure(figsize=(10,10))
sns.heatmap(data_corr.corr(), linewidths=0.1, vmax=1.0, square=True, cmap=colormap, annot=True)
plt.show()
```

#### Conclusions: there is a strong correlation between clusters. We will work with this in the future

```python
df = df.set_index(['user_id', 'post_id'])

df.head(50)
```

## Starting to teach

```python
### Let's start with a relatively simple model, and finish with boosting.
### How will we validate the data in time because randomly splitting the sample will not be correct.

max(df.timestamp), min(df.timestamp)
```

```python
df_train = df[df.timestamp < '2021-12-15']
df_test = df[df.timestamp >= '2021-12-15']

df_train = df_train.drop('timestamp', axis=1)
df_test = df_test.drop('timestamp', axis=1)

X_train = df_train.drop('target', axis=1)
X_test = df_test.drop('target', axis=1)

y_train = df_train['target']
y_test = df_test['target']

y_train.shape, y_test.shape
```

```python
df.head(50)
```

```python
### Let's define real features for processing them using OneHotEncoder or MeanTargetEncoder
object_cols = [
    'topic', 'TextCluster', 'gender', 'country',
    'city', 'exp_group', 'hour', 'month',
    'os', 'source'
]

cols_for_ohe = [x for x in object_cols if X_train[x].nunique() < 5]
cols_for_mte = [x for x in object_cols if X_train[x].nunique() >= 5]

### Let's save the indexes of these columns

cols_for_ohe_idx = [list(X_train.columns).index(col) for col in cols_for_ohe]
cols_for_mte_idx = [list(X_train.columns).index(col) for col in cols_for_mte]

t = [
    ('OneHotEncoder', OneHotEncoder(), cols_for_ohe_idx),
    ('MeanTargetEncoder', TargetEncoder(), cols_for_mte_idx)
]

col_transform = ColumnTransformer(transformers=t)

model_tree_classifier = Pipeline([("column_transformer",
                     col_transform),
                     
                    ("decision_tree", 
                     DecisionTreeClassifier())])

model_tree_classifier.fit(X_train, y_train)
```

## Let's look at both ROC-AUC and f1 score

```python
### Let's measure the quality of work of such a model
### Take ROC-AUC

print(f"Качество на трейне: {roc_auc_score(y_train, model_tree_classifier.predict_proba(X_train)[:, 1])}")
print(f"Качество на тесте: {roc_auc_score(y_test, model_tree_classifier.predict_proba(X_test)[:, 1])}")
```

```python
y_pred_train = np.around(model_tree_classifier.predict(X_train))  # predictions for the X-trainer
f1_score_train = np.array(f1_score(y_train, y_pred_train, average='weighted'))

y_pred_test = np.around(model_tree_classifier.predict(X_test))  # predictions on the X-test
f1_score_test = np.array(f1_score(y_test, y_pred_test, average='weighted'))
print(f"Quality on the train: {f1_score_train.round(3)}")
print(f"Quality on the test: {f1_score_test.round(3)}")
```

## Let's look at the ROC-AUC chart

```python
roc_auc_train_for_forest=[]
roc_auc_test_for_forest=[]

fpr_train, tpr_train, thresholds_train = roc_curve(y_train, model_tree_classifier.predict_proba(X_train)[:, 1])
fpr_test, tpr_test, thresholds_test = roc_curve(y_test, model_tree_classifier.predict_proba(X_test)[:, 1]) 
### On the train
roc_auc_train_for_forest.append(round(auc(fpr_train, tpr_train), 3))

### On the test
roc_auc_test_for_forest.append(round(auc(fpr_test, tpr_test), 3))
RocCurveDisplay.from_estimator(model_tree_classifier, X_train, y_train) # draw a graph for train
print(f"The ROC-AUC of a model tree classifier on the train is equal to {max(roc_auc_train_for_forest)}")

### On the test
RocCurveDisplay.from_estimator(model_tree_classifier, X_test, y_test) # draw a graph for the test
print(f"The ROC-AUC of a model tree classifier on the test is equal to {max(roc_auc_test_for_forest)}")
```

Conclusions: the decision tree was retrained on training data, and on test data the model is no better than a "coin toss" - 50/50

## Let's train the catboost model

```python
### Now let's train the catboost!

catboost = CatBoostClassifier(iterations=100,
                              learning_rate=1,
                              depth=2)

catboost.fit(X_train, y_train, object_cols)
```

## Let's look at both ROC-AUC and f1 score

```python
### Let's measure the quality of work of such a model
### Take ROC-AUC

print(f"Качество на трейне: {roc_auc_score(y_train, catboost.predict_proba(X_train)[:, 1])}")
print(f"Качество на тесте: {roc_auc_score(y_test, catboost.predict_proba(X_test)[:, 1])}")
```

```python
y_pred_train = np.around(catboost.predict(X_train))  # predictions for the X-trainer
f1_score_train = np.array(f1_score(y_train, y_pred_train, average='weighted'))

y_pred_test = np.around(catboost.predict(X_test))  # predictions on the X-test
f1_score_test = np.array(f1_score(y_test, y_pred_test, average='weighted'))
print(f"Quality on the train: {f1_score_train.round(3)}")
print(f"Quality on the test: {f1_score_test.round(3)}")
```

## Let's look at the ROC-AUC chart

```python
fpr_train_cboost, tpr_train_cboost, thresholds_train_cboost = roc_curve(y_train, catboost.predict_proba(X_train)[:, 1]) 
fpr_test_cboost, tpr_test_cboost, thresholds_test_cboost = roc_curve(y_test, catboost.predict_proba(X_test)[:, 1]) 

### On the train
RocCurveDisplay.from_estimator(catboost, X_train, y_train) # draw a graph for train
print(f"The ROC-AUC of CatBoost on the train is equal to {round(auc(fpr_train_cboost, tpr_train_cboost),3)}")

### On the test
RocCurveDisplay.from_estimator(catboost, X_test, y_test) # draw a graph for the test
print(f"The ROC-AUC of CatBoost on the test is equal to {round(auc(fpr_test_cboost, tpr_test_cboost), 3)}")
```

```python
y_pred_train = np.around(catboost.predict(X_train))  # predictions for the X-trainer
f1_score_train = np.array(f1_score(y_train, y_pred_train, average='weighted'))

y_pred_test = np.around(catboost.predict(X_test))  # predictions on the X-test
f1_score_test = np.array(f1_score(y_test, y_pred_test, average='weighted'))
print(f"Quality on the train: {f1_score_train.round(3)}")
print(f"Quality on the test: {f1_score_test.round(3)}")
```

Conclusions: the model is not retrained either on test or training data.

```python
### Let's look at feature_importance

def plot_feature_importance(importance,names,model_type):
    
    #Create arrays from feature importance and feature names
    feature_importance = np.array(importance)
    feature_names = np.array(names)
    
    #Create a DataFrame using a Dictionary
    data={'feature_names':feature_names,'feature_importance':feature_importance}
    fi_df = pd.DataFrame(data)
    
    #Sort the DataFrame in order decreasing feature importance
    fi_df.sort_values(by=['feature_importance'], ascending=False,inplace=True)
    
    #Define size of bar plot
    plt.figure(figsize=(10,8))
    #Plot Searborn bar chart
    sns.barplot(x=fi_df['feature_importance'], y=fi_df['feature_names'])
    #Add chart labels
    plt.title(model_type + 'FEATURE IMPORTANCE')
    plt.xlabel('FEATURE IMPORTANCE')
    plt.ylabel('FEATURE NAMES')
    
plot_feature_importance(catboost.feature_importances_,X_train.columns,'Catboost')
```
